4. 각 지점에서 판매량이 150개 이상인 데이터를 추출하고 그에 해당하는 지점과 과일을 출력하세요.

````python
import numpy as np

# 5개 지점의 월별 사과, 바나나, 오렌지 판매량 (5x3x6 배열)
# 5개의 지점, 3개의 과일(사과, 바나나, 오렌지), 6개월치 데이터
sales_data = np.array([
    # 지점 1
    [[120, 150, 200, 230, 180, 210],  # 사과 판매량
     [100, 140, 170, 190, 160, 180],  # 바나나 판매량
     [90, 120, 160, 140, 130, 150]],  # 오렌지 판매량

    # 지점 2
    [[80, 90, 110, 140, 130, 120],
     [60, 70, 90, 100, 110, 95],
     [50, 80, 100, 110, 90, 85]],

    # 지점 3
    [[130, 140, 160, 180, 170, 190],
     [110, 120, 130, 150, 140, 160],
     [100, 110, 120, 130, 120, 140]],

    # 지점 4
    [[100, 130, 150, 170, 160, 180],
     [90, 100, 110, 120, 130, 125],
     [80, 110, 140, 130, 120, 110]],

    # 지점 5
    [[90, 110, 130, 140, 120, 130],
     [70, 85, 100, 110, 105, 95],
     [60, 90, 110, 100, 90, 95]]
])

# 인덱싱을 위한 라벨 배열 (월 데이터도 배열로 만듭니다)
fruits = np.array(['사과', '바나나', '오렌지'])
branches = np.array(['지점 1', '지점 2', '지점 3', '지점 4', '지점 5'])
months = np.array(['1월', '2월', '3월', '4월', '5월', '6월'])
````

````python
# 1. 각 지점에서 6개월 동안 판매된 **사과**, **바나나**, **오렌지**의 월별 평균 판매량을 계산하세요.
avg_sales = np.mean(sales_data, axis=2)
# 지점 배열과 평균 배열을 가로로 결합 (column_stack)
result_1 = np.column_stack((branches, avg_sales))
print("[지점명, 사과, 바나나, 오렌지]")
print(result_1)
````

````python
# 2. 각 월마다 전체 지점에서 가장 많이 판매된 과일을 찾아 판매량과 과일명을 출력하세요.
monthly_total = np.sum(sales_data, axis=0)
max_idx = np.argmax(monthly_total, axis=0)
max_val = np.max(monthly_total, axis=0)
# 월, 1등과일, 판매량을 가로로 묶어서 한 번에 출력
result_2 = np.column_stack((months, fruits[max_idx], max_val))
print("[월, 과일명, 판매량]")
print(result_2)
````

````python
# 3. 6개월 동안 전체 지점에서 가장 많이 판매된 과일의 총량을 과일별로 구하세요. 어떤 과일이 가장 많이 판매되었는지 확인하세요.
total_sales = np.sum(sales_data, axis=(0, 2))
# 과일과 총판매량을 가로로 결합
result_3 = np.column_stack((fruits, total_sales))
print("[과일명, 총판매량]")
print(result_3)
print(f"👉 전체 1등 과일: {fruits[np.argmax(total_sales)]}")
````

````python
# 4. 각 지점에서 판매량이 150개 이상인 데이터를 추출하고 그에 해당하는 지점과 과일을 출력하세요.
# np.where로 조건에 맞는 튜플 반환 (지점_idx, 과일_idx, 월_idx)
b_idx, f_idx, m_idx = np.where(sales_data >= 150)
# 팬시 인덱싱으로 라벨을 한 번에 뽑아내고, 판매량 데이터도 한 번에 추출하여 결합
result_4 = np.column_stack((branches[b_idx], months[m_idx], fruits[f_idx], sales_data[sales_data >= 150]))
print("[지점, 월, 과일명, 판매량]")
print(result_4)
````
