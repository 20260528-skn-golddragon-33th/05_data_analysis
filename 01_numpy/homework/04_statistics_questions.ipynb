### [실습 1] 중심 경향 및 산포 측도 계산


아래 배열에서 평균, 중앙값, 최빈값, 분산, 표준편차, 범위를 계산하세요.


```python
arr = np.array([10, 20, 30, 30, 40, 50, 60])
```

````python
import numpy as np

arr = np.array([10, 20, 30, 30, 40, 50, 60])
print(arr)

print(np.mean(arr))
print(np.median(arr))
print(np.var(arr))
print(np.std(arr))
````

### [실습 2] 정규분포와 균등분포의 난수 생성


정규분포와 균등분포에서 각각 난수를 생성하여, 평균과 표준편차를 계산하세요.

1. 평균 0, 표준편차 1인 정규분포 난수 1000개 생성
2. 0과 1 사이의 균등분포 난수 1000개 생성

````python
# 1. 평균 0, 표준편차 1인 정규분포 난수 1000개 생성
arr1 = np.random.normal(loc=0.0, scale=1.0, size=1000)
print(np.mean(arr1))
print(np.std(arr1))

# 2. 0과 1 사이의 균등분포 난수 1000개 생성
arr2 = np.random.uniform(low=0.0, high=1.0, size=1000)
print(np.mean(arr2))
print(np.std(arr2))
````

### [실습 3] 평균 성적 구하기


아래와 같이 학생별 과목 성적 데이터가 주어졌을 때, 각 학생의 전체 평균 성적(수학, 영어, 과학)을 계산한 후, 이 평균값에 대한 분산과 표준편차를 구하세요.


```python
# 학생별 과목 성적 데이터
scores_math = [88, 92, 76, 85, 95, 92, 88, 90, 85, 88]
scores_english = [90, 85, 80, 87, 92, 89, 83, 91, 86, 88]
scores_science = [85, 88, 82, 89, 94, 86, 91, 88, 84, 87]
```

````python
scores_math = [88, 92, 76, 85, 95, 92, 88, 90, 85, 88]
scores_english = [90, 85, 80, 87, 92, 89, 83, 91, 86, 88]
scores_science = [85, 88, 82, 89, 94, 86, 91, 88, 84, 87]

scores_math = np.array(scores_math)
scores_english = np.array(scores_english)
scores_science = np.array(scores_science)

avg_score = np.mean([scores_math, scores_english, scores_science], axis=0)
print("분산: ", np.var(avg_score))
print("표준편차: ", np.std(avg_score))
````
